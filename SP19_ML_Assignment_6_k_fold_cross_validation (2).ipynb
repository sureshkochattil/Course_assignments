```python
import pandas as pd
import math
import random
from sklearn import tree

#load tae dataset 
ta_eval_raw = pd.read_csv("C:\\Users\\sures\\Downloads\\tae_data.csv")

#Randomizing dataset so that we don't have deterministic results (i.e. results can vary from run to run)

ta_eval_raw = ta_eval_raw.sample(frac=1)

#Adding column-names to dataset for readability and usability
ta_eval_raw.columns = ["N","I","C","S","P","Class"]


#k-fold cross validation
k=10

chunk_size = math.floor(ta_eval_raw.shape[0]/k)
accuracy_accumulation = []

for iteration in range(1, k+1):
    print("Computing fold where test chunk is:", k-iteration)
    
    #Ascribing the test_set and training_set depending on the fold
    test_set = ta_eval_raw.iloc[(k-iteration)*chunk_size:(k-iteration+1)*chunk_size, ]
    training_set_left = ta_eval_raw.iloc[:(k-iteration)*chunk_size, ]
    training_set_right = ta_eval_raw.iloc[(k-iteration+1)*chunk_size:, ]
    training_set = training_set_left.append(training_set_right)
    
    ta_eval_X = training_set.iloc[:, :5]
    ta_eval_Y = training_set.iloc[:, 5:6]
    
    decision_tree = tree.DecisionTreeClassifier(min_samples_leaf = 2, max_depth = 20)
    decision_tree = decision_tree.fit(ta_eval_X, ta_eval_Y)
      
    
    ta_eval_feature_names = list(ta_eval_X.columns.values)
    ta_eval_Y_Classes = ['1','2','3']
    
    true_value_training = list(training_set.iloc[:,5])
    true_value_test = list(test_set.iloc[:,5])
    
    prediction_on_trainingdata = decision_tree.predict(training_set.iloc[:, :5])
    prediction_on_testdata = decision_tree.predict(test_set.iloc[:, :5])
    
    predicted_values = []
    predicted_values.append(prediction_on_testdata)
    true_values = test_set[["Class"]]
    comparison_list = (true_values == predicted_values).values.T.tolist()
    accuracy = round(sum(comparison_list[0])/chunk_size,2)
    print("Fold Accuracy:", round(accuracy*100,2), "%")
    accuracy_accumulation.append(accuracy)
    
#Printing accuracy for each iteration
print(accuracy_accumulation)
#Printing averaged accuracy
print("Averaged Accuracy:", round(sum(accuracy_accumulation)/len(accuracy_accumulation)*100,2), "%")
#Printing averaged error
print("Averaged Error:", round((1-sum(accuracy_accumulation)/len(accuracy_accumulation))*100,2), "%")

   
```
