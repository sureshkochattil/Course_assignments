```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
from numpy import array
from sklearn.datasets import load_iris

import pandas as pd

iris = load_iris()

iris_dataframe = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
print(iris_dataframe.head())
```

```python
linear_reg = lm.LinearRegression()
```

```python
sepal_length = iris_dataframe[['sepal length (cm)']]
```

```python
sepal_width = iris_dataframe[['sepal width (cm)']]
```

```python
plt.plot(sepal_length,sepal_width,'bo')
plt.title("Sepal length vs Sepal width")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.show()
```

```python
linear_reg.fit(sepal_length,sepal_width)
```

```python
print("Sepal width = m*Sepal length + b")
print("Coefficients:",linear_reg.coef_)
print("Intercept:",linear_reg.intercept_)
print("Sepal width = -0.057 * Sepal length + 3.39")
```

```python
plt.plot(sepal_length,sepal_width,'bo')
plt.plot(sepal_length,linear_reg.predict(sepal_length), color = 'green', linewidth = 2)
plt.title("Sepal length vs Sepal width and Line of Fit")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.show()
```
