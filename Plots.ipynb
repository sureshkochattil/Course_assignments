```python
# Question # 3

from scipy.misc import imread
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# the file is the state of the union address 

filepath= "C:\\Users\\sures\\OneDrive\\Documents\\Trum_SOU.txt"

# Read the whole text.

def wordCloud(path):

    text = open(path).read() #read the entire file in one go

    wordcloud = WordCloud().generate(text)

    # Display the generated image:

    # the matplotlib way:

    plt.imshow(wordcloud)

    plt.axis("off")

    # take relative word frequencies into account, lower max_font_size

    wordcloud = WordCloud(background_color="white", max_words=2000,max_font_size=40, relative_scaling=.4).generate(text)

    plt.figure()

    plt.imshow(wordcloud)

    plt.axis("off")

    plt.show()


wordCloud(filepath)
```
