```python
# Import `DenseVector`
from pyspark.ml.linalg import DenseVector

# Define the `input_data` 
input_data = df.rdd.map(lambda x: (x[91], DenseVector(x[0:89])))

# Replace `df` with the new DataFrame
df = spark.createDataFrame(input_data, ["label", "features"])
#df["label"]
```

```python
from pyspark.ml.feature import MinMaxScaler
from pyspark.ml.linalg import Vectors


scaler = MinMaxScaler(inputCol="features", outputCol="features_scaled")

# Compute summary statistics and generate MinMaxScalerModel
scalerModel = scaler.fit(df)

# rescale each feature to range [min, max].
scaled_df = scalerModel.transform(df)

# Inspect the result
scaled_df.take(2)

#print("Features scaled to range: [%f, %f]" % (scaler.getMin(), scaler.getMax()))
#scaled_df.select("features", "features_scaled").show()
```

```python
train_data, validation_data, test_data = scaled_df.randomSplit([.7,.1,.2],seed=1234)
```

```python
# Import `LinearRegression`
from pyspark.ml.regression import LinearRegression

# Initialize `lr`
lr = LinearRegression(labelCol="label", maxIter=10, regParam=0.3, elasticNetParam=0.8)

# Fit the data to the model
linearModel = lr.fit(train_data)
```

```python
predicted = linearModel.transform(validation_data)

# Extract the predictions and the "known" correct labels
predictions = predicted.select("prediction").rdd.map(lambda x: x[0])
labels = predicted.select("label").rdd.map(lambda x: x[0])

# Zip `predictions` and `labels` into a list
predictionAndLabel = predictions.zip(labels).collect()
predictionAndLabel[:10]


#predictions = model.predict(testData.map(lambda x: x.features))
#labelsAndPredictions = testData.map(lambda lp: lp.label).zip(predictions)
#testErr = labelsAndPredictions.filter(
    #lambda lp: lp[0] != lp[1]).count() / float(testData.count())
#print('Test Error = ' + str(testErr))
#print('Learned classification GBT model:')
#print(model.toDebugString())
```
