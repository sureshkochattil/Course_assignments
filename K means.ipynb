```python
#Importing dataset
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#Dataset used
#https://archive.ics.uci.edu/ml/datasets/Sales_Transactions_Dataset_Weekly


#Reading Data and Alotting Column Names
sales_data = pd.read_csv("Sales_Transactions_Dataset_Weekly_t.csv")


#turning  as categorical and numerical

sales_data['Product_Code'] = sales_data['Product_Code'].astype('category')
sales_data['Product_Code'] = sales_data['Product_Code'].cat.codes
```

```python
plt.plot(sales_data['Product_Code'], sales_data['AVG_YEAR'], 'k.')
plt.title("Data Before K-Means")
plt.xlabel("Product Code")
plt.ylabel("Average Sale")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=100, tol=0.0001,precompute_distances='auto', verbose=0, random_state=None, copy_x=True, n_jobs=None, algorithm='auto')
```

```python
kmeans.fit(sales_data)
```

```python
y_kmeans = kmeans.predict(sales_data)
```

```python
plt.scatter(sales_data['Product_Code'], sales_data['AVG_YEAR'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=1);
plt.title("K-Means Results")
plt.xlabel("Product Code")
plt.ylabel("Average Sale")
plt.show()
```
